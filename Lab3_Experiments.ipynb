```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('rel_errors.xlsx')
df = df.sort_values(by=['Relative Error'])
```

```python
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.8
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'

fig, ax = plt.subplots(figsize=(10,5))
plt.hlines(y=df['Meas Tool'], xmin=0, xmax=df['Relative Error'], color='#007acc', alpha=0.2, linewidth=5)
plt.plot(df['Relative Error'], df['Meas Tool'], "o", markersize=5, color='#007acc', alpha=0.6)
ax.set_xlabel('Relative Error [%]', fontsize=15, fontweight='black', color = '#333F4B')
ax.set_ylabel('')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
plt.tight_layout()
plt.show()
fig.savefig('relative_plot.png')
```

```python
import seaborn as sns
```

```python
df2 = pd.read_excel('res_uncer.xlsx')
df2
```

```python
sns.set(style="whitegrid")


# Draw a pointplot to show pulse as a function of three categorical factors
g = sns.catplot(x="flow[%]", y="Q[m^3/sec]", col="Type",# hue="Type",
                capsize=.2, palette="YlGnBu_d", height=6, width=8, aspect=.75, col_wrap=3,
                kind="point", data=df2)
g.set_xticklabels([34,53 , 71])
g.set(ylim=(9, 36))
g.despine(left=True)

plt.savefig('new.png')
```

```python
plt.close()
```
